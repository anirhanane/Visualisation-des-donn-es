```python
def add_categorical_legend(folium_map, title, colors, labels):
    if len(colors) != len(labels):
        raise ValueError("colors and labels must have the same length.")

    color_by_label = dict(zip(labels, colors))
    
    legend_categories = ""     
    for label, color in color_by_label.items():
        legend_categories += f"<li><span style='background:{color}'></span>{label}</li>"
        
    legend_html = f"""
    <div id='maplegend' class='maplegend'>
      <div class='legend-title'>{title}</div>
      <div class='legend-scale'>
        <ul class='legend-labels'>
        {legend_categories}
        </ul>
      </div>
    </div>
    """
    script = f"""
        <script type="text/javascript">
        var oneTimeExecution = (function() {{
                    var executed = false;
                    return function() {{
                        if (!executed) {{
                             var checkExist = setInterval(function() {{
                                       if ((document.getElementsByClassName('leaflet-top leaflet-right').length) || (!executed)) {{
                                          document.getElementsByClassName('leaflet-top leaflet-right')[0].style.display = "flex"
                                          document.getElementsByClassName('leaflet-top leaflet-right')[0].style.flexDirection = "column"
                                          document.getElementsByClassName('leaflet-top leaflet-right')[0].innerHTML += `{legend_html}`;
                                          clearInterval(checkExist);
                                          executed = true;
                                       }}
                                    }}, 100);
                        }}
                    }};
                }})();
        oneTimeExecution()
        </script>
      """
   

    css = """

    <style type='text/css'>
      .maplegend {
        z-index:9999;
        float:right;
        background-color: rgba(255, 255, 255, 1);
        border-radius: 5px;
        border: 2px solid #bbb;
        padding: 10px;
        font-size:12px;
        positon: relative;
      }
      .maplegend .legend-title {
        text-align: left;
        margin-bottom: 5px;
        font-weight: bold;
        font-size: 90%;
        }
      .maplegend .legend-scale ul {
        margin: 0;
        margin-bottom: 5px;
        padding: 0;
        float: left;
        list-style: none;
        }
      .maplegend .legend-scale ul li {
        font-size: 80%;
        list-style: none;
        margin-left: 0;
        line-height: 18px;
        margin-bottom: 2px;
        }
      .maplegend ul.legend-labels li span {
        display: block;
        float: left;
        height: 16px;
        width: 30px;
        margin-right: 5px;
        margin-left: 0;
        border: 0px solid #ccc;
        }
      .maplegend .legend-source {
        font-size: 80%;
        color: #777;
        clear: both;
        }
      .maplegend a {
        color: #777;
        }
    </style>
    """

    folium_map.get_root().header.add_child(folium.Element(script + css))

    return folium_map
```

```python
#on rajoute un fond de carte
carte = folium.Map(location=[46.5, 2.3], zoom_start=6)
pv=ds_pointVente()

for i in pv.index:
    html=f"""
        <h4>Nom: {pv["nom"][i]}</h4>
        <h5>Type: {pv["type"][i]}</h5>
        <h5>Bls: {pv["bls"][i]}</h5>
        """
    iframe = folium.IFrame(html=html, width=200, height=200)
    popup = folium.Popup(iframe, max_width=2650)
    loc=[pv.iloc[i]['coordinates'][1], pv.iloc[i]['coordinates'][0]]
   
    if ((pv.iloc[i]['type'] == 'GARE' or pv.iloc[i]['type'] == 'Gare') and (pv.iloc[i]['bls'] == 'oui' or pv.iloc[i]['bls'] == 'Oui') ):
        folium.Marker(loc, popup = popup ,icon=folium.Icon(color='green')).add_to(carte)
    if ((pv.iloc[i]['type'] == 'GARE' or pv.iloc[i]['type'] == 'Gare') and (pv.iloc[i]['bls'] == 'non' or pv.iloc[i]['bls'] == 'Non') ):
        folium.Marker(loc, popup = popup,icon=folium.Icon(color='blue')).add_to(carte)
    if (pv.iloc[i]['type'] == 'BOUTIQUE' and (pv.iloc[i]['bls'] == 'Non' or pv.iloc[i]['bls'] == 'non')):
        folium.Marker(loc, popup = popup,icon=folium.Icon(color='orange')).add_to(carte)
    if (pv.iloc[i]['type'] == 'BLS SEULE' and (pv.iloc[i]['bls'] == 'oui' or pv.iloc[i]['bls'] == 'Oui')):
        folium.Marker(loc, popup = popup,icon=folium.Icon(color='red')).add_to(carte)
carte =add_categorical_legend(carte, 'Type',
                             colors = ['green','blue','orange','red'],
                           labels = ['Gare, bls oui','Gare, bls non', 'Boutique, bls non','Bls seule , bls oui'])
plugins.Fullscreen(
    position="topleft",
    title="Expand me",
    title_cancel="Exit me",
    force_separate_button=True,
).add_to(carte)
carte
```

```python
def preparation_pd1(deter):
    data = deter
    data = data[data['taux_de_ponctualite'] > 0]
    data1 = data
    data1 = data1.assign(dateF=data1['date'].astype('datetime64[ns]'))
    data1
    data11 = data1.assign(regionG=data1['region'])
    data11['regionG'] = data11['regionG'].apply(lambda x: 'Auvergne-Rhône-Alpes' if x == 'Auvergne' else x)
    data11['regionG'] = data11['regionG'].apply(lambda x: 'Auvergne-Rhône-Alpes' if x == 'Auvergne-Rhône Alpes' else x)
    data11['regionG'] = data11['regionG'].apply(lambda x: 'Auvergne-Rhône-Alpes' if x == 'Rhône Alpes' else x)
    data11['regionG'] = data11['regionG'].apply(lambda x: 'Bourgogne-Franche-Comté' if x == 'Franche Comté' else x)
    data11['regionG'] = data11['regionG'].apply(lambda x: 'Bourgogne-Franche-Comté' if x == 'Bourgogne' else x)
    data11['regionG'] = data11['regionG'].apply(lambda x: 'Bourgogne-Franche-Comté' if x == 'Bourgogne-Franche Comté' else x)
    data11['regionG'] = data11['regionG'].apply(lambda x: 'Occitanie' if x == 'Midi Pyrénées' else x)
    data11['regionG'] = data11['regionG'].apply(lambda x: 'Occitanie' if x == 'Languedoc Roussillon' else x)
    data11['regionG'] = data11['regionG'].apply(lambda x: 'Hauts de France' if x == 'Picardie' else x)
    data11['regionG'] = data11['regionG'].apply(lambda x: 'Hauts de France' if x == 'Nord Pas de Calais' else x)
    data11['regionG'] = data11['regionG'].apply(lambda x: 'Centre Val-de-Loire' if x == 'Centre' else x)
    data11['regionG'] = data11['regionG'].apply(lambda x: 'Grand Est' if x == 'Champagne Ardenne' else x)
    data11['regionG'] = data11['regionG'].apply(lambda x: 'Grand Est' if x == 'Lorraine' else x)
    data11['regionG'] = data11['regionG'].apply(lambda x: 'Grand Est' if x == 'Alsace' else x)
    data11['regionG'] = data11['regionG'].apply(lambda x: 'Nouvelle Aquitaine' if x == 'Aquitaine' else x)
    data11['regionG'] = data11['regionG'].apply(lambda x: 'Nouvelle Aquitaine' if x == 'Limousin' else x)
    data11['regionG'] = data11['regionG'].apply(lambda x: 'Nouvelle Aquitaine' if x == 'Poitou Charentes' else x)
    data11['regionG'] = data11['regionG'].apply(lambda x: 'Normandie' if x == 'Haute Normandie' else x)
    data11['regionG'] = data11['regionG'].apply(lambda x: 'Normandie' if x == 'Basse Normandie' else x)
    data11['regionG'] = data11['regionG'].apply(lambda x: 'Pays de la Loire' if x == 'Pays-de-la-Loire' else x)
    data11['regionG'] = data11['regionG'].apply(lambda x: 'Hauts de France' if x == 'Hauts-de-France' else x)
    data11 = data11.assign(mois=data11['dateF'].dt.month)
    data11 = data11[data11['regionG']!='Sud']
    dataa = data11[data11['date'].str.split("-").str.get(0).astype(int) <= 2019]
    dataa
    data_t = data11[data11['date'].str.split("-").str.get(0).astype(int) > 2019]
    data_t
    return data11,dataa, data_t
```
